## Масштабирование — это просто изменение размера изображения

```python
import numpy as np
import cv2 as cv

img = cv.imread('C:/Users/sanya/Downloads/cK1tuAJa6j8.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"

res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)
```

```python
print(res.shape)
print(img.shape)
```

```python
cv.imshow('img',res)
k = cv.waitKey(0)
```

```python
height, width = img.shape[:2]
res2 = cv.resize(img,(2*width, 2*height), interpolation = cv.INTER_CUBIC)

print(res2.shape)
```

```python
cv.imshow('img',res2)
k = cv.waitKey(0)
```

## изменение местоположения объекта

```python
img = cv.imread('C:/Users/sanya/Downloads/cK1tuAJa6j8.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv.warpAffine(img,M,(cols,rows))

cv.imshow('img',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Поворот изображения на угол

```python
img = cv.imread('C:/Users/sanya/Downloads/cK1tuAJa6j8.jpg', cv.IMREAD_GRAYSCALE)
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols = img.shape

# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
dst = cv.warpAffine(img,M,(cols,rows))

cv.imshow('img',dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Аффинное преобразование

```python
import matplotlib.pyplot as plt

img = cv.imread('C:/Users/sanya/Downloads/cK1tuAJa6j8.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv.getAffineTransform(pts1,pts2)

dst = cv.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

## Трансформация перспективы

```python
img = cv.imread('C:/Users/sanya/Downloads/sudoky.jpeg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv.getPerspectiveTransform(pts1,pts2)

dst = cv.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
